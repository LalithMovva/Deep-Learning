```python
a=[]
b=[]
c=[]
d=[]
e=[]
f=[]

for i in range(0,50):
    a=a+ [df['PATHLENGTH'][i]]
    b=b+ [df['InitialTrust'][i]]
    c=c+ [df['TransitionThreshold'][i]]
    d=d+ [df['No of Ants'][i]]
    e=e+ [df['ACCURACY'][i]]
    f=f+ [df['AVERAGE PATH'][i]]
```

```python
df8 = pd.DataFrame(
{
    "pa":a,
    "it":b,
    "tt":c,
    "na":d,  
})
```

```python
df8
```

```python
df2 = pd.DataFrame(
{
    'acc':e,
    'ap':f
})
```

```python
from sklearn.model_selection import train_test_split
```

```python
df2
```

```python
x = df8.to_numpy()
```

```python
y = df2.to_numpy()
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=[4]),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32,activation='relu'),
    tf.keras.layers.Dense(2)
  ])

optimizer = tf.keras.optimizers.RMSprop(0.001)

model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
```

```python
model.summary()
```

```python
hist=model.fit(
  x_train, y_train,
  epochs=1000)
```
